```python
# remove N/A and infinite values
import numpy as np 
df2_theory=df2_theory.replace(np.inf,None)
df2_theory=df2_theory.replace(-np.inf,None)
df2_theory=df2_theory.replace('na',None)

df3=df2_theory.dropna(subset=IV)
len(df3)
```

```python
# Split the dataset into training and testing 
TestYear=list(range(2015, 2018))

dff={}
for each in TestYear:
    Test=df3.loc[df3['Data Year - Fiscal']==each]  
    Test['Label']='Test'
    
    Train=df3.loc[df3['Data Year - Fiscal'].isin(list(range(each-6, each-1)))]
    Train['Label']='Train'
    
    Set=pd.concat([Train, Test], ignore_index=True,sort=False) 
    dff['dff_{}'.format(each)]=Set
```

```python
# Pre-processing by standardizing the continuous variables 
def Transform(X):
    for name in list(Cont_IV):
        X[name]=(X[name]-X[name].mean())/X[name].std()
        
        X=X[['CIK Number','Data Year - Fiscal','RES','Label']+IV]
        
        X=X.replace(np.inf,None)
        X=X.replace(-np.inf,None)
        X=X.replace('na',None)
        
        X=X.dropna(subset=IV)
        
        
    return X
```

```python
# Use the 20 most predictive ARVs
best20features=['IS_NTH_ADD_OP',
'Abs(Accruals)',
                'Log Audit Report Lag',
                'Abs(Accruals/CFO)',
'NOTEFF_OTHER_x',
'Industry Specialization (sale)_national',
'Influence',
'Tenure_x',
'Industry Specialization (sale)_city',
'DD Residual',
'AUDIT_FEES',
'WorkloadCompression',
'Office Size_sales',
'AbnormalLAF',
'Disc. Accruals', 
               'COMBINED_IC_OP',
'MATERIAL_WEAKNESS',
'Abs (Disc. Accruals)',
'GOING_CONCERN',
'IS_EFFECTIVE']

len(best20features)
```

```python
predictors=best20features

AUC=[]
BestModel=[]
Test_Year=[]
Cost=[]


testyears=list(range(2015, 2018)) 
costratios=[1, 10, 20, 30, 40, 50]

for TestYear in testyears:
    df=dff['dff_{}'.format(TestYear)]
    trans_df=Transform(df)
    
    #train and test for final results
    test=trans_df[trans_df['Data Year - Fiscal']==TestYear]
    train=trans_df[trans_df['Data Year - Fiscal']<=TestYear-2]

    #split training data into training for validation and testing for validation
    vali_test=trans_df[trans_df['Data Year - Fiscal']==TestYear-2]
    vali_train=trans_df[trans_df['Data Year - Fiscal']<=TestYear-4]

    test_X=test[predictors]
    test_y=test['RES']
    train_X=train[predictors]
    train_y=train['RES']

    vali_test_X=vali_test[predictors]
    vali_test_y=vali_test['RES']
    vali_train_X=vali_train[predictors]
    vali_train_y=vali_train['RES']

    #Tune Hyperparameters of XGB
    param_grid = {'eta': [0.01, 0.2, 0.3], 
       'min_child_weight': [1, 5],
      'max_depth': [3, 6, 10],
        'subsample':[0.5, 1],
        'colsample_bytree':[0.5, 1]}

    h1=[]
    h2=[]
    h3=[]
    h4=[]
    h5=[]

    Algorithm=[]

    Tune_AUC=[]

    vali_train_RES_indices = vali_train[vali_train.RES==1].index  #get the index of RES instances in vali_train 
    vali_train_NORES_indices = vali_train[vali_train.RES==0].index #get the index of NON-RES instances in vali_train 
    
    for num_split in costratios:
    
        splitted_NORES=np.array_split(np.random.permutation(vali_train_NORES_indices),num_split) #randomly split the NON-RES instances in vali-train into num_split batches

        for H1 in param_grid['eta']:
            for H2 in param_grid['min_child_weight']:
                for H3 in param_grid['max_depth']:
                    for H4 in param_grid['subsample']:
                        for H5 in param_grid['colsample_bytree']:
        
                            tune=XGBClassifier(eta=H1, 
                                                min_child_weight=H2, 
                                                max_depth=H3,
                                                subsample=H4,
                                                colsample_bytree=H5) #for model with each and every combination of hyperparameters
                            y_score_array=[] # to store probability predictions from models trained by each subsample
                            for splitted_NORES_indices in splitted_NORES: # here we adopt the OU method from Perols et al. (2017)
                                under_sample_indices = np.concatenate([vali_train_RES_indices,splitted_NORES_indices]) # for each batch, create an undersampled training set
                                vali_train_under_sample_y = vali_train_y.loc[under_sample_indices]
                                vali_train_under_sample_X = vali_train_X.loc[under_sample_indices]

                                probas_=tune.fit(vali_train_under_sample_X, vali_train_under_sample_y).predict_proba(vali_test_X)
                                y_score=probas_[:, 1]
                                y_score_array.append(y_score)

                            #average the prediction probabilities from different batches        
                            y_average_score=[]
                            for i in range(len(vali_test_y)):
                                average_score=np.mean(y_score_array, axis=0)[i]
                                y_average_score.append(average_score)


                            auc=roc_auc_score(vali_test_y, y_average_score)
                            Tune_AUC.append(auc)

                            h1.append(H1)
                            h2.append(H2)
                            h3.append(H3)
                            h4.append(H4)
                            h5.append(H5)



        HyperTune_Results=pd.DataFrame()
        HyperTune_Results['h1']=h1
        HyperTune_Results['h2']=h2
        HyperTune_Results['h3']=h3
        HyperTune_Results['h4']=h4
        HyperTune_Results['h5']=h5

        HyperTune_Results['AUC']=Tune_AUC
    
        tunedresults=HyperTune_Results[HyperTune_Results['AUC']==HyperTune_Results['AUC'].max()].reset_index() #The tuned model is the one that maximizes out-of-sample AUC on the hold-out validation for tuning data

                        
        bestmodel=XGBClassifier(eta=tunedresults['h1'][0], 
                                     min_child_weight=tunedresults['h2'][0], 
                                    max_depth=tunedresults['h3'][0],
                                    subsample=tunedresults['h4'][0],
                                    colsample_bytree=tunedresults['h5'][0]) #use the hyperparameters selected above


        #Use the best model to make classification
        train_RES_indices = train[train.RES==1].index  
        train_NORES_indices =train[train.RES==0].index  

        splitted_NORES=np.array_split(np.random.permutation(train_NORES_indices),num_split)

        y_score_array=[] # used to collect prediction probability from each split 
        y_average_score=[] # will store average prediction probability 
       
        for splitted_NORES_indices in splitted_NORES: # here we adopt the OU method from Perols et al. (2017)
            under_sample_indices = np.concatenate([train_RES_indices,splitted_NORES_indices]) # for each batch, create an undersampled training set
            train_under_sample_y = train_y.loc[under_sample_indices]
            train_under_sample_X = train_X.loc[under_sample_indices]

            probas_ = bestmodel.fit(train_under_sample_X, train_under_sample_y).predict_proba(test_X) 
            #obtain the probability prediction 
            y_score=probas_[:, 1]
            y_score_array.append(y_score)

        num_test=len(test_y) 
        for i in range(num_test):
            average_score=np.mean(y_score_array, axis=0)[i] # take the average of the probability scores from all splits
            y_average_score.append(average_score)


        auc=roc_auc_score(test_y, y_average_score)
        AUC.append(auc)

        Test_Year.append(TestYear)
        Cost.append(num_split)
        BestModel.append(bestmodel)

    

results=pd.DataFrame()
results['Test Year']=Test_Year
results['AUC']=AUC
results['Cost']=Cost
results['BestModel']=BestModel
results
```
